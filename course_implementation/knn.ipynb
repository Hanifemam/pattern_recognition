```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/Users/hanifemamgholizadeh/Desktop/patter_recognition/data/iris_knn_dataset.csv')
data.head()
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
data.head()
```

```python
test_data = list(data.iterrows())[-1]
print(test_data)
```

```python
data = data.drop(data.index[-1])
data.tail()
```

```python
def calculate_distance(row1, row2, columns, p=2):
    raw_distance = 0
    for column in columns:
      raw_distance += np.abs((row1[column] - row2[column]) ** p)
      print(raw_distance ** (1/p))
    return raw_distance ** (1/p)
```

```python
distance_to_test = []
data_train = data.drop(columns=['target'])
for row in data.iterrows():
    distance_to_test.append(calculate_distance(row[1], test_data[1], data_train.columns))
    
len(distance_to_test), distance_to_test[:5]
```

```python
def three_min_indexes(lst):
    if len(lst) < 3:
        raise ValueError("List must contain at least three elements.")
    
    min_vals = [(float('inf'), -1)] * 3  # (value, index)

    for idx, num in enumerate(lst):
        if num < min_vals[0][0]:
            min_vals[2] = min_vals[1]
            min_vals[1] = min_vals[0]
            min_vals[0] = (num, idx)
        elif num < min_vals[1][0] and idx != min_vals[0][1]:
            min_vals[2] = min_vals[1]
            min_vals[1] = (num, idx)
        elif num < min_vals[2][0] and idx != min_vals[0][1] and idx != min_vals[1][1]:
            min_vals[2] = (num, idx)
    
    return [i[1] for i in min_vals]
```

```python
print(distance_to_test)
three_least_distance = three_min_indexes(distance_to_test)
print(three_least_distance)
```

```python
import statistics
predicted_classes = list(data.iloc[three_least_distance]['target'].values)
mode = statistics.mode(predicted_classes)
print("Predicted class:", mode)
```

```python

data_features = data.drop(columns=['target'])
data_target = data['target']
X_train, X_test, y_train, y_test = train_test_split(data_features, data_target, test_size=0.2, random_state=42)
```

```python
preds = []
for _, features in X_test.iterrows():
    distance_to_test = []
    for _, row in X_train.iterrows():
        distance_to_test.append(calculate_distance(row, features, columns=X_train.columns))
    three_least_distance = three_min_indexes(distance_to_test)
    predicted_classes = list(y_train.iloc[three_least_distance].values)
    mode = statistics.mode(predicted_classes)
    preds.append(mode)
    
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, preds)
accuracy
```
