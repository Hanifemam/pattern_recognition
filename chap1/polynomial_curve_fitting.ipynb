```python
from scipy.stats import stats
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_size = 100
x_values = [2 * np.pi * random.random() for _ in range(sample_size)]
y_values = np.sin(x_values) + (np.random.normal(loc=0, scale=0.1, size=sample_size))
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)  
```

```python
x_values
```

```python
y_values
```

```python
sns.scatterplot(x=x_values, y=y_values)
sns.lineplot(x=x, y=y, color="green")
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.01, 1.01)
```
