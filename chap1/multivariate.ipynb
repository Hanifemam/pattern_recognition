```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
## Calculation of Gaussian Distribution
```

```python
mu = np.array([[0], [0]])
sigma = np.array([[1, 0.5], [0.5, 1]])
x = np.array([[1], [1]])
def gaussian(x, mu, sigma):
    d = len(mu)
    coeff = 1 / ((2 * np.pi) ** (d / 2) * np.sqrt(np.linalg.det(sigma)))
    exponent = -0.5 * ((x - mu).T @ np.linalg.inv(sigma) @ (x - mu))
    return coeff * np.exp(exponent)
```

## Simulation phase

```python
x1_range = np.linspace(-5, 5, 100)
x2_range = np.linspace(-5, 5, 100)
x1, x2 = np.meshgrid(x1_range, x2_range)
gaussian_distribution = np.zeros_like(x1)
for i in range(100):
    for j in range(100):
        x_vec = np.array([[x1[i, j]], [x2[i, j]]])
        gaussian_distribution[i, j] = gaussian(x_vec, mu, sigma)
        

```

## Drawing the simulation graph

```python


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x1, x2, gaussian_distribution, cmap='viridis')

ax.set_zlim(0, 0.2)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of Multivariate Gaussian')

plt.show()
```
